```python
import csv
```

```python
# Read in raw data from csv
rawData = csv.reader(open('lga_2017.csv', 'r'), dialect='excel')
line_count = len(list(csv.reader(open('lga_2017.csv', 'r'), dialect='excel')))
```

```python
# the template. where data from the csv will be formatted to geojson
template = \
    ''' \
    { 
      "type": "Feature",
      "geometry": {
            "type": "Point",
            "coordinates": [%s, %s]
      },
      "properties": {
            "id": %s,      
            "LGA": "%s",
            "Grand_Total": %s,
            "Humanitarian": %s,
            "Family": %s,
            "Skilled": %s
      }
    }'''
```

```python
# insert comma between objects
comma = \
''',
'''
```

```python
# the head of the geojson file
output = \
    ''' \
{ 
    "type": "FeatureCollection",
    "features": [
    '''
```

```python
# loop through the csv by row skipping the first
iter = 0
for row in rawData:
    iter += 1
    if (iter >= 2)&(row[8] != 'Bad_address'):
        id = row[0]
        LGA = row[7]
        lat = row[8]
        lon = row[9]
        Grand_Total = row[5]
        Humanitarian = row[2]
        Family = row[3]
        Skilled = row[4]
        # output += template % (row[0], row[2], row[1], row[3], row[4])
        output += template % (lon, lat, id, LGA, Grand_Total, Humanitarian, Family, Skilled)
        if iter < line_count:
            output += comma
```

```python
# the tail of the geojson file
output += \
    ''' \
    ]
}
    '''
```

```python
# opens an geoJSON file to write the output to
outFileHandle = open("lga_2017.geojson", "w")
outFileHandle.write(output)
outFileHandle.close()
```
