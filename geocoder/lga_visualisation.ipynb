```python
import pandas as pd
import folium
```

```python
lga_2017 = pd.read_csv('lga_2017.csv')
```

```python
lga_2017.head(9)
```

```python
m = folium.Map(location=[-25, 135], zoom_start=4)
```

```python
#Iterate through the list of LGAs and add circle markers on a map

row_counter=0

for row in lga_2017.itertuples():
    if row.Lat != 'Bad_address':
        lga = row.LGA
        lat = float(row.Lat)
        lon = float(row.Lng)
        grand_total = int(row.Grand_Total)
        popup = folium.Popup('LGA: '+lga+'. Growth:'+str(grand_total), parse_html=True)
        folium.Circle(
            radius=grand_total,
            location=[lat, lon],
            popup=popup,
            color='blue',
            fill=True,
            parse_html=True,
        ).add_to(m)
    row_counter += 1
```
